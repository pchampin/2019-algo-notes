```python
def palindrome(txt: str) -> bool:
    """
    post-cond: retourne True ssi txt est un palindrome
    """
    est_palindrome = True
    i = 0
    while i<len(txt)//2 and est_palindrome:
        est_palindrome = txt[i]==txt[len(txt)-1-i]
        i = i+1
    return est_palindrome


# version alternative :

def palindrome(txt: str) -> bool:
    """
    post-cond: retourne True ssi txt est un palindrome
    """
    est_palindrome = True
    i = 0
    j = len(txt)-1  # symétrique de i dans txt
    while i<j and est_palindrome:
        est_palindrome = txt[i]==txt[j]
        i = i+1
        j = j-1
    return est_palindrome
```

```python
def intercale(txt1: str, txt2: str) -> str:
    """
    post-cond: retourne une chaine alternant les caractères de txt1 et txt2
               en commençant par txt1,
               et en remplaçant les caractères "manquants" par des espaces
    """
    
    résultat = ""
    if len(txt1) > len(txt2):
        n = len(txt1)
    else:
        n = len(txt2)
    
    i = 0
    while i<n:
        if i<len(txt1):
            c1 = txt1[i]
        else:
            c1 = " "
        if i<len(txt2):
            c2 = txt2[i]
        else:
            c2 = " "
        résultat = résultat+c1+c2
        i = i+1
    return résultat


# version alternative :

def intercale(txt1: str, txt2: str) -> str:
    """
    post-cond: retourne une chaine alternant les caractères de txt1 et txt2
               en commençant par txt1,
               et en remplaçant les caractères "manquants" par des espaces
    """
    
    résultat = ""
    # si txt1 est plus courte que txt2, on la complète avec des espaces
    while len(txt1) < len(txt2):
        txt1 = txt1+" "
    # si txt2 est plus courte que txt1, on la complète avec des espaces
    while len(txt2) < len(txt1):
        txt2 = txt2+" "
    # maintenant, les 2 chaînes ont la même longueur
    i = 0
    while i<len(txt1):
        résultat = résultat+txt1[i]+txt2[i]
        i = i+1
    return résultat
```

```python
def contient(txt: str, sub: str) -> bool:
    """
    post-cond: retourne True ssi txt contient sub
    """
    trouvé = False
    i = 0
    while i <= len(txt)-len(sub) and not trouvé:
        correspond = True
        j = 0
        while j < len(sub) and correspond:
            correspond = txt[i+j]==sub[j]
            j = j+1
        trouvé = correspond
        i = i+1
    return trouvé
```
